```python
# import in and read indiana_census.csv and 2016_US_County_Level_Presidential_Results.csv files
import pandas as pd
import requests
import json
import csv

election_results_file = "in_clean_result.csv"
indiana_census_file = "indiana_census.csv"
indiana_census2_file = "indiana_census2.csv"

election_results_df = pd.read_csv(election_results_file)
indiana_census_df = pd.read_csv(indiana_census_file)
indiana_census2_df = pd.read_csv(indiana_census2_file)
```

```python
election_results_df.head()
```

```python
# Clean up county_name category. Replace 'county_name' with 'County'
election_results_df.rename(columns={"county_name":"County"}, inplace=True)
election_results_df.columns
```

```python
# Merge two dataframes using an outer join
merged_file_1 = pd.merge(election_results_df, indiana_census_df, on="County", how="outer")
```

```python
# Merge last dataframe and first merged file dataframes using an outer join
merged_file2 = pd.merge(merged_file_1, indiana_census2_df, on="County", how="outer")
```

```python
merged_file2.head()
```

```python
merged_file2.columns
```

```python
# Drop columns that are not needed
merged_file3 = merged_file2.drop(["Unnamed: 0_x", "Unnamed: 0_y", "State_y"], axis = 1)
```

```python
# Rename columns
merged_file3.rename(columns=
                   {"County":"County",
                    "State_x":"State",
                   "votes_dem":"DEM Votes",
                   "votes_gop":"GOP Votes",
                   "total_votes":"Total Votes",
                   "per_dem":"DEM %",
                   "per_gop":"GOP %",
                   "state_abbr":"State Abbr",                   
                   "Total Population":"Total Population",
                   "Population (White)":"Population (White)",
                   "Population (African-American)":"Population (African-American)",
                   "Population (Asian)":"Population (Asian)",
                   "Population (Hispanic)":"Population (Hispanic)",
                   "Median Home Value":"Median Home Value",
                   "Education (None)":"Education (None)",
                   "Education (High School)":"Education (High School)",
                   "Education (GED)":"Education (GED)",
                   "Education (Associates)":"Education (Associates)",
                   "Education (Bachelors)":"Education (Bachelors)",
                   "Education (Masters)":"Education (Masters)",
                   "Education (Professional)":"Education (Professional)",
                   "Education (Doctorate)":"Education (Doctorate)",
                   "Median Income":"Median Income",
                   "Income Per Capita":"Income Per Capita",
                   "Pop in Labor Force":"Pop in Labor Force",
                   "Pop not in Labor Force":"Pop Not in Labor Force",
                   "Unemployed":"Unemployed",
                   "Median Age":"Median Age",
                   "Median Age (male)":"Median Age (Male)",
                   "Median Age (female)":"Median Age (Female)"},
                   inplace=True)
```

```python
merged_file3.columns
```

```python
# Reorganizing the columns using double brackets
merged_file = merged_file3[['State', 'State Abbr','County', 'DEM Votes', 'GOP Votes', 'Total Votes', 'DEM %', 'GOP %', 
       'Total Population', 'Population (White)', 'Population (African-American)', 'Population (Asian)',
       'Population (Hispanic)', 'Median Age', 'Median Age (Male)', 'Median Age (Female)',
        'Education (None)', 'Education (High School)', 'Education (GED)', 'Education (Associates)',
       'Education (Bachelors)', 'Education (Masters)','Education (Professional)', 'Education (Doctorate)',
        'Median Income', 'Income Per Capita', 'Median Home Value',
        'Pop in Labor Force', 'Pop Not in Labor Force', 'Unemployed', ]]
merged_file.head()
```
