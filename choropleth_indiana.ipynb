```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import plotly.figure_factory as ff
```

```python
# The path to our CSV file
csv_path = "2016_US_County_Level_Presidential_Results.csv"
result_df = pd.read_csv(csv_path)

#result_df.head()
in_result = result_df[result_df['state_abbr'] == 'IN']
in_clean_result = in_result.drop(['Unnamed: 0', 'diff', 'per_point_diff'], axis=1)
```

```python
def getmethewinner(row):
    if row['votes_dem'] >= row['votes_gop']:
        val = 1
    else:
        val = 2
    return val
```

```python
in_clean_result['winner'] = in_clean_result.apply(getmethewinner, axis=1)
in_clean_result['dem_percent'] = round(in_clean_result.per_dem*100, 0)
#in_clean_result = in_clean_result.loc[in_clean_result.winner == 1]
in_clean_result
```

```python
values = in_clean_result['dem_percent'].tolist()
fips = in_clean_result['combined_fips'].tolist()

endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = ["#67001f","#67001f","#053061","#053061","#053061","#053061","#053061","#053061"]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['Indiana'], show_state_data=True,
    colorscale=colorscale, binning_endpoints=endpts, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Winner by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    exponent_format=True,
)
fig.layout.template = None
fig.show()
```

```python
import plotly.express as px
```

```python
geojson = px.data.election_geojson()

fig = px.choropleth(in_clean_result, geojson=geojson, color="winner" , scope=['usa'], 
                    locations="county_name", featureidkey="properties.district",
                    projection="mercator", hover_data=["winner"]
                   )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
